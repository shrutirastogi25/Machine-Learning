# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Graph

```python
x=np.arange(20)
print(x)
plt.plot(x)
plt.show()
```

## Scatter Plot Graphs

```python
y=np.array([1,2,3,4,5])
```

```python
z=y**2
plt.scatter(y,z,color="orange",label="Squares", marker="o")
plt.legend()
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.show()
```

```python
plt.plot(x,x**2, label="Squares",marker="o")
plt.plot(x,x**3, label="Cubes", marker="o")
plt.legend()
plt.xlabel("Numbers")
plt.ylabel("Squares and Cubes")
plt.show()
```

## Bar Graphs

```python
numbersA=np.random.randint(0,10,5)
indicesA=np.arange(5)
indicesA=indicesA+2015
numbersB=np.random.randint(0,10,5)
indicesB=np.arange(5)
indicesB=indicesB+2015
print(numbersA)
print(numbersB)
```

```python
plt.bar(indicesA,numbersA,0.25,color="blue",label="Rainfall")
plt.bar(indicesB+0.25,numbersB,0.25,color="red",label="Humidity")
plt.legend()
plt.show()
```

## Pie Chart

```python
subjects=["english","hindi","maths","science","social science","computers"]
marks=[90,80,40,73,78,43]
plt.pie(marks,labels=subjects,radius=1)
plt.show()
plt.pie(marks,labels=subjects,radius=1,explode=(0,0,1,0,0,0),shadow=True)
plt.show()
```

## Histogram

```python
u=5
sigma=2
vals=u+sigma*np.random.randn(1000)
print(vals.shape)
plt.hist(vals,50)
plt.show()
```
