# Plots and Charts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,4*np.pi,0.1)
```

## Sine function

```python
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

## Cosine Function

```python
z = np.cos(x)
plt.plot(x,z)
plt.show()
```

## Labeling in the Graphs

```python
plt.plot(x,y,x,z)
plt.xlabel('values of x')
plt.ylabel('sin(x) and cos(x)')
plt.title('Sine and Cosine Function')
plt.legend(['sin(x)', 'cos(x)'])
plt.show()
```
